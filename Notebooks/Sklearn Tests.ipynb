###Example 1

This is the first example that was given for the mixture package for sklearn. It generates 2 random sets of Gaussian data and fits to it.

```python
import numpy as np
from sklearn import mixture
from matplotlib import pyplot as plt
%matplotlib inline
from IPython.core.pylabtools import figsize
figsize(12,10)

n_samples = 300
np.random.seed(0)
C = np.array([[0,-.7],[1.5, .7]]) #the distorion matrix for the 1st sample

#1st sample is distored by C
sample1 = np.dot(np.random.randn(n_samples,2),C)
#2nd samples is translated 20 right and 20 up
sample2 = np.random.randn(n_samples,2) + np.array([20,20])

X_train = np.r_[sample1, sample2] #combine to one array

#Create a Gaussian Mixure Model with 2 components and full covariance.
clf = mixture.GMM(n_components = 2, covariance_type = 'full')
#There are 4 types of covariance: full, diag, sphere, and tied. Full is the most free, but occasionally overfits

clf.fit(X_train)#the actual fit

x = np.linspace(-20.0, 30.0)
y = np.linspace(-20.0, 40.0)
X,Y = np.meshgrid(x, y)
XX = np.c_[X.ravel(), Y.ravel()]

Z = np.log(-clf.score_samples(XX)[0])#log likelihood of the data
Z = Z.reshape(X.shape)

#plotting stuff
CS = plt.contour(X,Y,Z)
CB = plt.colorbar(CS, shrink = .8, extend = 'both')

plt.scatter(X_train[:,0], X_train[:,1], .8)

plt.axis('tight')
```

That was pretty easy, and it did a good job. Let's see what happens when we try to break it. First, let's mess with the number of components we specify. 

```python
#change the num of objects to 3 instead of 2
clf = mixture.GMM(n_components = 3, covariance_type = 'full')

#Everything below here is copied from above
clf.fit(X_train)#the actual fit

x = np.linspace(-20.0, 30.0)
y = np.linspace(-20.0, 40.0)
X,Y = np.meshgrid(x, y)
XX = np.c_[X.ravel(), Y.ravel()]

Z = np.log(-clf.score_samples(XX)[0])#log likelihood of the data
Z = Z.reshape(X.shape)

#plotting stuff
CS = plt.contour(X,Y,Z)
CB = plt.colorbar(CS, shrink = .8, extend = 'both')

plt.scatter(X_train[:,0], X_train[:,1], .8)

plt.axis('tight')
```

So apparently there is a really strong dependance on the number of components. That could be damning, but there are 2 other packages included in the mixture module. There are the DPGMM and VBGMM packages as well. There are details about there differences here: http://scikit-learn.org/stable/modules/mixture.html

The VBGMM module is like GMM, but uses some other algorithm to make apporximations, and apparently works better when there is a smaller amount of data. DPGMM is interesting because it only takes as many components as it needs, and takes the passed in specification as a maximum. For example:

```python
#Now 10 instead of 2
clf = mixture.DPGMM(n_components = 10, covariance_type = 'full')

#Everything below here was copied from above. 
clf.fit(X_train)#the actual fit

x = np.linspace(-20.0, 30.0)
y = np.linspace(-20.0, 40.0)
X,Y = np.meshgrid(x, y)
XX = np.c_[X.ravel(), Y.ravel()]

Z = np.log(-clf.score_samples(XX)[0])#log likelihood of the data
Z = Z.reshape(X.shape)

#plotting stuff
CS = plt.contour(X,Y,Z)
CB = plt.colorbar(CS, shrink = .8, extend = 'both')

plt.scatter(X_train[:,0], X_train[:,1], .8)

plt.axis('tight')
```

###Example 2
The second example is pretty expansive. It goes about selecting the best model out of several. I'll reproduce the whole example, and then force it to stop working afterwards.

```python
import itertools
from scipy import linalg

n_samples = 500
C = np.array([[0, -.1],[1.7, .4]])
sample1 = np.dot(np.random.randn(n_samples,2),C)
sample2 = .7*np.random.randn(n_samples,2) + np.array([-6,3])

X = np.r_[sample1, sample2]

lowest_bic = np.infty
bic = [] #BIC stands for Bayesian Information Criterion, it's a meassure of the goodness of a fit

n_components_range = range(1,7)
for cv_type in cv_types:
    for n_components in n_components_range:
        #Fit with this specifc number of components and covariance
        gmm = mixture.GMM(n_components = n_components, covariance_type = cv_type)
        gmm.fit(X)
        bic.append(gmm.bic(X))
        #if this BIC is lower, this model is better
        if bic[-1]<lowest_bic:
            lowest_bic= bic[-1]
            best_gmm = gmm
            
bic = np.array(bic)
color_iter = itertools.cycle(['k','r','g','b','c','m','y'])

clf = best_gmm
bars = []

#all plotting below here. Probably just skip this. 
spl = plt.subplot(211)
for i,(cv_type,color) in enumerate(zip(cv_types, color_iter)):
    xpos = np.array(n_components_range) + .2*(i-2)
    bars.append(plt.bar(xpos, bic[i*len(n_components_range):
                                (i+1)*len(n_components_range)],
                      width = .2, color = color))
plt.xticks(n_components_range)
plt.ylim([bic.min()*1.01-.01*bic.max(), bic.max()])
plt.title('BIC score per model')
xpos = np.mod(bic.argmin(), len(n_components_range)) + .65+\
                .2*np.floor(bic.argmin()/len(n_components_range))
plt.text(xpos, bic.min()*.97+.03*bic.max(), '*', fontsize = 14)
spl.set_xlabel('Number of components')
spl.legend([b[0] for b in bars], cv_types)

splot = plt.subplot(212)
Y_ = clf.predict(X) #this is what labels as a point from one of the many gaussians

from matplotlib import patches
for i, (mean,covar,color) in enumerate(zip(clf.means_, clf.covars_, color_iter)):
    v,w = linalg.eigh(covar)
    if not np.any(Y_==i):
        continue
    plt.scatter(X[Y_==i,0], X[Y_==i, 1],.8, color = color)

    angle = np.arctan2(w[0][1], w[0][0])
    angle = 180*angle/np.pi
    v*=4
    ell = patches.Ellipse(mean, v[0], v[1], 180+angle, color = color)
    ell.set_clip_box(splot.bbox)
    ell.set_alpha(.5)
    splot.add_artist(ell)

plt.xlim(-10, 10)
plt.ylim(-3, 6)
plt.xticks(())
plt.yticks(())
plt.title('Selected GMM: full model 2 components')
plt.subplots_adjust(hspace = .35, bottom = .02)
```
