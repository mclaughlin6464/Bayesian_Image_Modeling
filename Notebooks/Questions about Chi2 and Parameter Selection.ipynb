###Emcee Code for Fitting Images
So I'd been prioritizing multinest over emcee for the last few weeks but after some experimenting I think emcee will ultimately be better.  Emcee offers a lot more control, and is much much faster with increasing number of gaussians. 

```python
import numpy as np
%matplotlib inline
from IPython.core.pylabtools import figsize
figsize(12,8)

#my standard easy to use gaussian function
def gaussian(x,y,cx,cy,a,r):
    return a*np.exp(-1./(r**2)*((x-cx)**2+(y-cy)**2))

#log prior
def lnprior(theta):
    #theta contains the variables for amplitude and width
    #theta = [A1, A2, ...AN, R1, ...RN]
    N = len(theta)/2
    amps = theta[:N]
    rads = theta[N:]
    #these uniform priors are for the exponents of amplitdue and radius
    if not all(0<a<3 for a in amps):
        return -np.inf
    if not all(-1<r<1 for r in rads):
        return -np.inf
    return 0

#log liklihood
def lnlike(theta, image, xx, yy, c_x, c_y, inv_sigma2):
    N = len(theta)/2
    amps = theta[:N]
    rads = theta[N:]
    
    model = sum(gaussian(xx,yy,c_x,c_y,10**a,10**r) for a,r in zip(amps,rads))
    
    return -.5*(np.sum(((image-model)**2)*inv_sigma2-np.log(inv_sigma2)))

#log probability
def lnprob(theta, image, xx, yy, c_x, c_y, inv_sigma2):
    lp = lnprior(theta)
    if np.isfinite(lp):
        return lp+lnlike(theta, image, xx, yy, c_x, c_y, inv_sigma2)
    return -np.inf
```

The above is all pretty standard for what I've done before. Normally,  I'd do all the fitting in the global scope. This time, however, I've written it as a function, so it can be called multiple times with different parameters.

```python
import emcee as mc
from multiprocessing import cpu_count
from collections import Counter
from scipy.stats import chisquare
from time import time
def fit_image(image, N, c_x=-1, c_y = -1, n_walkers = 600):
    #seed the random number generator
    np.random.seed(int(time()))
    
    if -1 in (c_x,c_y): #at least one is the default
        #If no center given, use the image maximum
        c_y,c_x = np.unravel_index(image.argmax(), image.shape)
        
    boxRatio = .05 #the size of the box around the galaxy of interest
    
    img_y, img_x = image.shape
    
    xLims = [c_x+img_x*boxRatio*(-1)**(1+i) for i in xrange(2)]
    yLims = [c_y+img_y*boxRatio*(-1)**(1+i) for i in xrange(2)]
    
    #trim the image down around the galaxy of interest
    image = image[yLims[0]:yLims[1], xLims[0]:xLims[1]]
    
    #reassign these parameters now that we've cropped the image 
    img_y, img_x = image.shape
    c_y,c_x = c_y-yLims[0], c_x-xLims[0]
    
    y,x = [np.arange(i) for i in image.shape]
    xx,yy = np.meshgrid(x,y) #useful for speedier computation
    
    err = .1
    inv_sigma2 = 1./(err**2) #necessary for calculation. Value doesn't seem to be importatn
    
    ndim = N*2
    nburn = n_walkers*.1
    nsteps = 200
    
    #initial random guess
    pos = []
    for walk in xrange(n_walkers):
        row = np.zeros(ndim)
        for n in xrange(N):
            row[n] = 3*np.random.rand()
            row[n+N] = 2*np.random.rand()-1
        pos.append(row)
        
    args = (image, xx, yy, c_x, c_y, inv_sigma2)
    
    sampler = mc.EnsembleSampler(n_walkers, ndim, lnprob, args = args, threads = cpu_count())
    
    #run the sampler
    sampler.run_mcmc(pos, nsteps)
    #get the samples and split up into amps and rads
    samples = sampler.chain[:,nburn:,:].reshape((-1, ndim))
    sampled_as = samples[:, :N].reshape((-1))
    sampled_rs = samples[:, N:].reshape((-1))
    
    #I'll explain the use of these more in greater detail
    Amp_counter = Counter(np.round(sampled_as, decimals = 1).tolist())
    Rad_counter = Counter(np.round(sampled_rs, decimals = 1).tolist())
    
    calc_as ,a_counts = zip(*Amp_counter.most_common(N))
    calc_rs, r_counts = zip(*Rad_counter.most_common(N))
    #the calculated fit
    calc_img = sum(gaussian(xx,yy,c_x,c_y,10**a, 10**r) for a,r in zip(calc_as, calc_rs))
    
    #perform a chi2 test on the fit
    chi2stat, b = chisquare(image, f_exp = calc_img, axis = None) 
    
    return image, calc_img, chi2stat
```

###The Chi Squared Test
So above I use Scipy's chisquare test to assess the "goodness" of a fit. You might already be familiar with how it works, but in case you aren't I show a simple application of it below.

```python
data = np.array(range(1,11)) + np.random.normal(size =10)
fit = range(1,11)
print data
print fit
print chisquare(data, f_exp = fit)
```

So the first number in the tuple is the Chi2 statistic, and the second is the probability that the deviations from the fit are due to random chance. So for this example it works great but let's apply it to my fit.
